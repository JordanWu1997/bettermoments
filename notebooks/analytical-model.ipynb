```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from diskmodel import disk_model
```

```python
# build a disk model and convolve with a circular 0.3" beam.
axis, velax, data, vproj = disk_model(Nchan=64, Npix=128, beam=0.3)
print data.shape
```

```python
# plot the rotation profile.
fig, ax = plt.subplots()
im = ax.pcolor(axis, axis, vproj / 1e3)
cb = plt.colorbar(im, pad=0.02, ticks=np.arange(-6, 8, 2))
ax.set_aspect(1)
ax.set_xlim(axis.min(), axis.max())
ax.set_ylim(axis.min(), axis.max())
cb.set_label('Projected Rotation (m/s)', rotation=270, labelpad=15)
ax.set_xlabel('Offset (arcsec)')
ax.set_ylabel('Offset (arcsec)')
```

```python
# plot an example channel.
fig, ax = plt.subplots()
im = ax.pcolor(axis, axis, data[30])
cb = plt.colorbar(im, pad=0.02, ticks=np.arange(0, 30, 5))
ax.set_aspect(1)
ax.set_xlim(axis.min(), axis.max())
ax.set_ylim(axis.min(), axis.max())
cb.set_label('Brightness Temperatuer (K)', rotation=270, labelpad=15)
ax.set_xlabel('Offset (arcsec)')
ax.set_ylabel('Offset (arcsec)')
```
